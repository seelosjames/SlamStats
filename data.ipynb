```python
def createMatchup(index, box_info, matchup_info):
    matchup = {}

    matchup['matchup_id'] = index
    matchup['home_team'] = matchup_info['home_team']
    matchup['home_team_ranking'] = int(matchup_info['home_team_ranking'])
    matchup['home_team_box'] = index * 2
    matchup['away_team'] = matchup_info['home_team']
    matchup['away_team_ranking'] = int(matchup_info['home_team_ranking'])
    matchup['away_team_box'] = (index * 2) + 1
    matchup['url'] = matchup_info['box_score_url']
    matchup['neutral'] = matchup_info['neutral']

    return matchup
```

```python
def createScoring(index, box_info, matchup_info):
    scoring = {}

    scoring['matchup_id'] = index
    scoring['home_first'] = int(box_info['home_1'])
    scoring['home_second'] = int(box_info['home_2'])
    scoring['home_ot'] =  int(box_info['home_tot']) - (int(box_info['home_1']) + int(box_info['home_2']))
    scoring['home_total'] = int(box_info['home_tot'])
    scoring['away_first'] = int(box_info['home_1'])
    scoring['away_second'] = int(box_info['home_2'])
    scoring['away_ot'] = int(box_info['away_tot']) - (int(box_info['away_1']) + int(box_info['away_2']))
    scoring['away_total'] = int(box_info['away_tot'])

    return scoring
```
